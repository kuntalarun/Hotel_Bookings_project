3 **DATA WRANGLING**

```python
# let's check which variable have a null values.we have already some variables who's have null values!
missing_value[:4]
```

```python
#let's check what is the percentage of null  values in each columns .startin from company !
percentage_of_null = missing_value[0]/new_row*100
percentage_of_null
```

```python
# so now we have to drop this column 'company' because it's have a lot's of null values 
drop_column=df.drop(['company'],axis=1 ,inplace=True)
drop_column
```

```python
#now let's check for agent 
percentage_of_agent = missing_value[1]/new_row*100
percentage_of_agent
```

```python
#as we seen that agent as a some null values so we are fill that null values to'0' !
df['agent'].fillna(value=0,inplace=True)
df['agent'].isnull().sum()
```

```python
#let's check percentage to country column !
percentage_country=missing_value[2]/new_row*100
percentage_country
```

```python
# so now u can see column 'country' has a less amount of null values so now we have to change this into 'others'!
df['country'].fillna(value='others',inplace=True)
df['country'].isnull().sum()
```

```python
# now we are calculate the percentage of column "children"!
percentage_children=missing_value[3]/new_row*100
percentage_children
```

```python
# now we are change the values of children column to '0' !
df['children'].fillna(value=0 , inplace=True)
df['children'].isnull().sum()
```

```python
# let's check dataset have any other null values 
df.isnull().sum()# as we see that dataset have not any other null values
```

Change in datatype for some required column

```python
# first we check the type of dataset columns
df.info()
```

```python
#u can see 'children' and 'agent' type is float let's we have to change the type of 'children' and 'agent' column into int64 !
df[['children','agent']]=df[['children','agent']].astype('int64')
```

Addition of new columns as per requirement

```python
#let's find out total stay at night !
df['total_stay_at_night']=df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_stay_at_night']
```

```python
# now we are add the one extra col 'revenue' for using total_stay_at_night * adr !
df['revenue']=df['total_stay_at_night']*df['adr']
df['revenue']
```

```python
#now we add some columns for take total_guest coming for each booking!
df['total_guest'] = df['adults'] + df['children'] + df['babies']
df['total_guest'].sum()
```

```python
#for undrestanding the cl:'is_canceled' values repplace with(0,1) 'not_canceled' or ' canceled'!
df['is_canceled']=df['is_canceled'].replace([0,1],['not_canceled','is_canceled'])
df['is_canceled']
```

```python
# same as 'is_repeated_guest' column !
df['is_repeated_guest']=df['is_repeated_guest'].replace([0,1],['not_repeated','repeated'])
df['is_repeated_guest']
```

```python
# now we are check overall revenue hotel wise !
hotel_total_revenue = df.groupby('hotel')['revenue'].sum()
hotel_total_revenue
```
